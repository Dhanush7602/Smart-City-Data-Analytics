```python
# Importing necessary libraries
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
# Plot histograms for air_quality_df
#Histograms are used to visualize the distribution of data. They make it easy to see how often each range of values appears.
for column in air_quality_df.columns:
    if air_quality_df[column].dtype != 'object':  
        plt.figure(figsize=(8, 4))
        plt.hist(air_quality_df[column], bins=30, color='blue', edgecolor='black')
        plt.title(f'Distribution of {column} in Air Quality Dataset')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()
```

```python
# Plot histograms for infrastructure_df
for column in infrastructure_df.columns:
    if infrastructure_df[column].dtype != 'object':  # Only plot numerical columns
        plt.figure(figsize=(8, 4))
        plt.hist(infrastructure_df[column], bins=30, color='green', edgecolor='black')
        plt.title(f'Distribution of {column} in Infrastructure Dataset')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()
```

```python
# Plotting histograms for traffic_df
for column in traffic_df.columns:
    if traffic_df[column].dtype != 'object':  # Only plot numerical columns
        plt.figure(figsize=(8, 4))
        plt.hist(traffic_df[column], bins=30, color='orange', edgecolor='black')
        plt.title(f'Distribution of {column} in Traffic Dataset')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()
```

```python
#A correlation heatmap helps you understand how numerical variables in the dataset are related to each other
#we implement correlation heatmaps for all the three datasets from here
#1 means perfect positive correlation (both increase together).
#-1 means perfect negative correlation (one increases, the other decreases).
#0 means no correlation.
import seaborn as sns
import matplotlib.pyplot as plt

# Filter out non-numeric columns from the dataset
numeric_air_quality_df = air_quality_df.select_dtypes(include=['float64', 'int64'])  # Only select numeric columns

# Plot correlation heatmap for air_quality_df (only numeric columns)
plt.figure(figsize=(10, 8))
correlation_matrix_air_quality = numeric_air_quality_df.corr()  # Correlation matrix for numerical data only
sns.heatmap(correlation_matrix_air_quality, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Air Quality Dataset')
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Filtering out non-numeric columns from the infrastructure dataset 
numeric_infrastructure_df = infrastructure_df.select_dtypes(include=['float64', 'int64'])  # Only select numeric columns

# Plot correlation heatmap for infrastructure_df (only numeric columns)
plt.figure(figsize=(10, 8))
correlation_matrix_infrastructure = numeric_infrastructure_df.corr()  # Correlation matrix for numerical data only
sns.heatmap(correlation_matrix_infrastructure, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Infrastructure Dataset')
plt.show()
```

```python
# Filter out non-numeric columns from the traffic dataset
numeric_traffic_df = traffic_df.select_dtypes(include=['float64', 'int64'])  # Only select numeric columns

# Plot correlation heatmap for traffic_df (only numeric columns)
plt.figure(figsize=(10, 8))
correlation_matrix_traffic = numeric_traffic_df.corr()  # Correlation matrix for numerical data only
sns.heatmap(correlation_matrix_traffic, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Traffic Dataset')
plt.show()
```

```python
#Box Plots for Detecting Outliers 
#A box plot is a simple and effective way to detect outliers (extreme data points that differ significantly from the rest of the data).
#It visualizes the distribution of the data by quartiles and highlights outliers(outliers mean value that is completely different from 
#other values in a set )
# Box plot for air_quality_df to detect outliers
for column in air_quality_df.columns:
    if air_quality_df[column].dtype != 'object':  # Only for numerical columns
        plt.figure(figsize=(8, 4))
        sns.boxplot(air_quality_df[column])
        plt.title(f'Box Plot of {column} - Air Quality Dataset')
        plt.show()



```
